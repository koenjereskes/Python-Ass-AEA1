### 1. Abstract

Multinational firms with operations in high-tax countries can benefit the most from reallocating
taxable income to tax havens. By using the theoretical framework by [Gumpert, Hines and Schnitzer, 2016](http://cepr.org/sites/default/files/news/CEPR_FreeDP_120916.pdf), we estimate the domestic tax rate for multinationals where they will not report any domestic profits under different profit shifting costs. We find that the lower the costs, the lower the domestic tax rate needs to be in order for a firm to shift all their profits to a possible foreign affiliate.

### 2. Introduction

Tax havens are typically small countries with high-quality governance institutions. They impose low or zero tax rates on foreign investors ([Dharmapala and Hines, 2009](http://www.nber.org/papers/w12802.pdf)). Sophisticated tax avoidance strategies involving tax havens have received considerable attention in the media ([The Independent, 2016](http://www.independent.co.uk/news/uk/politics/tax-havens-bermuda-worst-zero-per-cent-corporation-tax-rate-a7469386.html)). There are roughly 40 major tax havens in the world today, but the sizable apparent economic returns to becoming a tax haven raise the question of why there are not more. Both the OECD and a dataset by Rice and Hines identify which tax havens there are <a href="http://www.nber.org/papers/w12802.pdf#page=34">Appendix Table 1 Dharmapala and Hines, 2009</a>. Tax havens have been in the focus of national and international policy measures. Their interest is in the concern that tax havens are widely used to avoid tax obligations in their high-tax countries. The purpose of our investigation is to identify the factors that determines the demand for tax havens and the incentive of profit shifting. According to [Gumpert, Hines and Schnitzer, 2016](http://cepr.org/sites/default/files/news/CEPR_FreeDP_120916.pdf), who focus their investigation on German multinationals, state that a one percent increase in the foreign tax rate leads to a 2.3 per cent increase in having a tax haven affiliate,  reflecting the greater benefit of reallocating taxable income from high-tax countries.

### 3. Theoretical framework

In this section we create the theoretical framework behind the paper of [Gumpert, Hines and Schnitzer, 2016](http://cepr.org/sites/default/files/news/CEPR_FreeDP_120916.pdf). The theoretical framework is based on the incentive for a multinational firm to invest in a tax haven.
A multinational firm can choose invest in a range of countries $i = 0,...,n.$ where country 0 is the taxhaven.
Other variables are:

$c_i$ = Fixed set-up cost

$\rho_i$ = Before-tax profits earned in country $i$ by the affiliate once installed

$\tau_i$ = reported profits are taxed at this rate in country $i$

Assume $\tau_0$ = 0 i.e there is no taxation in tax haven.

$\psi_i$ = amount of actual profits in country $i$ that can be reallocated to a country with lower tax rate.
Assume income reallocation gets increasingly expensive as the amount reallocated increases relative to income earned in country $i$.

$(\alpha/2)(\psi_i^2/\rho_i)$ are the assumed costs income reallocation according to <a href = "http://www.nber.org/papers/w3477.pdf" > Hines and Rice (1994)</a>. 

Where $\alpha$ is a parameter that captures how much the cost of income reallocation increases with the amount reallocated. Note that $a$ is firm specific and we will elaborate on this in more detail in our <a href='#Discussion'>discussion</a> session.
Now we assume that the cost of reallocation are incurred in the country from which income is shifted.
Reported profit in country $i$, $\pi_i$, after fixed cost $c_i$ are sunk is thus
\begin{eqnarray}
\pi_i = \rho_i - \psi_i - \frac{\alpha}{2} \frac{\psi_i^2}{\rho_i}
\end{eqnarray}

The investor's maximization problem, given that firm has a tax haven affiliate, is

\begin{equation}
\max_{d_i,\psi_i}\sum_{i=1}^{n} d_i \Bigg[\psi_i + (1-\tau_i)(\rho_i - \psi_i - \frac{\alpha}{2} \frac{\psi_i^2}{\rho_i})-c_i\Bigg]
\end{equation}

With $d_i$$\in$ {$0,1$}, s.t.

\begin{eqnarray}
\rho_i - \psi_i - \frac{\alpha}{2} \frac{\psi_i^2}{\rho_i} \geq 0 \space \forall_i = 1,...,n.
\end{eqnarray}

***LEMMA 1 : If the investor has a tax haven affiliate, the optimal amount of profit shifting is:***

\begin{eqnarray}
\psi_i^{*th} = \frac{\tau_i\rho_i}{\alpha(1-\tau_i)}
\end{eqnarray}

Which can be rewritten into:

\begin{eqnarray}
\tau_i = \frac{\psi_i^{*th}\rho_i}{\alpha(1+\psi_i^{*th}\rho_i)}
\end{eqnarray}

### 3.2. Execution<a id='Execution'></a>

```python
tag= 0.315 # tax rate for German firm
tah = 0.15 # Tax rate for Hungarian firm
rho = 10000 # Before tax profits earned in country i we assume 10000
alpha = 0.4 # Cost of income reallocation
psig= (tag*rho)/(alpha*(1-tag)) # amount of profit shifting German firm
psih = (tah*rho)/(alpha*(1-tah)) # amount of profit shifting Hungarian firm
print psig
print psih
```

```python
taug = (psig*alpha/rho)/(1+((psig*alpha)/rho)) #Domestic tax function for Germany
tauh = (psih*alpha/rho)/(1+((psih*alpha)/rho)) #Domestic tax function for Hungary
print taug
print tauh
```

```python
taus = np.arange(0, 0.6, 0.01);
psis = (taus*rho)/(alpha*(1-taus))
plt.plot(taus*100, psis, "b", label="a=0.4")
plt.axvline(x=taug*100,color="thistle",linestyle="--") # German firm
plt.axhline(y=psig,color="thistle",linestyle="--") # German firm
plt.axvline(x=tauh*100,color="thistle",linestyle="--") # Hungarian firm
plt.axhline(y=psih,color="thistle",linestyle="--") # Hungarian firm
plt.title("Domestic tax rate vs optimal amount of profit shifting")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel ('Domestic tax rate in percentages')
plt.ylabel ('Profit shifting in euros')
```

In this paper, the main focus is to determine the impact of tax rates on the incentive to set up a tax haven. Proposition 1 describes how a change in tax rates in countries in which the multinational is active independent of a tax haven investment, i.e $i=1,...,\hat{n},$ affects the incentive to set up a tax haven.

***Proposition 1<a id='Proposition'></a>: The larger the tax rate in countries in which the multinational is active independent of a tax haven investment, the larger is the incentive to establish a tax haven affiliate.***

\begin{eqnarray}
\frac{dInc_{th}}{d\tau_i} = \frac{d(\Pi_i^{th}-\Pi_i^{nth})}{d\tau_i}= (\psi_i^{*th}-\psi_i^{*nth}) + \Bigg[\frac{a}{2}\frac{\psi_i^{*th2}}{\rho_i} - \frac{a}{2}\frac{\psi_i^{*nth2}}{\rho_i}\Bigg] > 0  \space \forall_i = 1,...,\hat{n}.
\end{eqnarray}

This proposition will be empirically estimated later in the paper.The effect of changes in countries in which the multinational is active only in case of a tax haven investment.

We want to prove this proposition by using the following theoretical framework by [Gumpert, Hines and Schnitzer, 2016](http://cepr.org/sites/default/files/news/CEPR_FreeDP_120916.pdf):

\begin{eqnarray}
(1-\tau_i)\pi_i = (1-\tau_i)(\rho_i - \psi_i - \frac{a}{2} \frac{\psi_i^2}{\rho_i}) = (1-\tau_i) \bigg[1- \frac{\tau_i(2-\tau_i)}{2\alpha(1-\tau_i)^2}\bigg] \rho_i
\end{eqnarray}
Which can be rewritten to
\begin{eqnarray}
\pi_i = \bigg[1- \frac{\tau_i(2-\tau_i)}{2\alpha(1-\tau_i)^2}\bigg] \rho_i
\end{eqnarray}

```python
rho = 10000 # Before tax profits earned in country i we assume 10000
alpha = 0.4 #baseline
taup = np.arange(0, 0.6, 0.01); 

rp = (1-((taup*(2-taup))/(2*0.4*(1-taup)**2))) * rho # Formula where rp(reported profits) = œÄi
yline = np.arange(0, 0.6, 0.01) # In order to see the intersection with the y=0 where 0 profits are reported in the home country
plt.plot(taup*100,rp, "b",label="a=0.4") 
plt.plot(taup*100, yline , "y",label="y=0")
idx = np.argwhere(np.diff(np.sign(rp-yline)) != 0).reshape(-1) + 0 # In order to place a point on the intersection between rp and yline
plt.plot(taup[idx]*100, rp[idx], 'ko')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.axvline(x=25,color="thistle",linestyle="--") # For clearing the x-value of intersection even more.
plt.title("Unobservability of actual profits")
plt.xlabel ('Domestic tax rate in percentages')
plt.ylabel ('Reported profits in domestic country in euros')
```

### 5. Discussion<a id='Discussion'></a>
In our model the costs of setting up a tax haven affiliate $c_i$ are fixed, and increases with the size of reallocated profits relative to the total reallocatable income. Since only this ratio, and not absolute size of reallocated profits, increases $c_i$, firms with higher pre-tax profits should be more likely to reallocate taxable income. The data indicate that the larger and more productive the manufacturing firms are, the more likely they are to have tax haven affiliates[(Gumpert, 2016)](http://cepr.org/sites/default/files/news/CEPR_FreeDP_120916.pdf). Manufacturing firms with non-haven foreign operations in high-tax countries are more likely to have tax affiliates than other firms and especially more likely than firms in the service industry. Multinational firms with higher profits, and thus a higher tax burden, are more likely to reallocate taxable income.
In the model a firm specific parameter $alpha$ is included to allow for firm or sector specific tax haven investment. This can explain the empirical differences in tax haven investment between firms in for example manufacturing and the service industry. The relationship between tax rates and owning a tax haven affiliate is stronger for manufacturing firms than for  firms in service industries, possibly reflecting the difficulty of reallocating taxable service income. 

Also, it should be noted that these results apply only to tax systems that are similar to the German system. In a sample of German multinational firms of the period 2002-2008 only 20.4% have tax haven affiliates, which is considerably lower than in the US. Even when the sample is limited to only manufacturing firms, this difference persists [(Gumpert, 2016)](http://cepr.org/sites/default/files/news/CEPR_FreeDP_120916.pdf). This implies that cost effective tax avoidance for German firms is limited while it is more widespread in the US. This is despite the stricter US tax system, which does more to limit the benefits of reallocating taxable income than the German system. Under US tax law, foreign subsidiaries  of US firms have a tax liability on repatriated income, which encourages these subsidiaries to direct foreign profits to new foreign investments. The German tax system is similar to that of most OECD countries and taxes only 5 percent of the active foreign profits of its resident corporations. Therefore German companies have no incentives to not repatriate foreign earned income since this does not result in a higher tax burden. This results in incentives to establish tax haven affiliates that are different to those of US firms and similar to those of other G7 and OECD firms.
